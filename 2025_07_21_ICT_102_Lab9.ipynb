```python
import pandas as pd

# Load each CSV into a DataFrame
students_df = pd.read_csv('students.csv')
courses_df = pd.read_csv('courses.csv')
enrollments_df = pd.read_csv('enrollments.csv')
```

```python
# Display all DataFrames
print("ðŸ“˜ Students Table")
print(students_df, end="\n\n")

print("ðŸ“— Courses Table")
print(courses_df, end="\n\n")

print("ðŸ“™ Enrollments Table")
print(enrollments_df)

```

```python
from IPython.display import display

display(students_df)
display(courses_df)
display(enrollments_df)
```

```python
# demonstrate common relational database operations using your uploaded CSV files and pandas DataFrames in Google Colab.
#âœ… 1. Each DataFrame as a Table (Relation)
print("ðŸŽ“ Students Table")
print(students_df)

print("\nðŸ“š Courses Table")
print(courses_df)

print("\nðŸ“ Enrollments Table")
print(enrollments_df)
```

```python
# âœ… 2. SQL-Like Operations Using pandas
# ðŸ”¹ SELECT â†’ Filtering rows and columns

# Select students older than 20
selected_students = students_df[students_df['age'] > 20]
print("Students older than 20:\n", selected_students)

# Select only name and age columns
name_age_only = students_df[['name', 'age']]
print("\nNames and Ages:\n", name_age_only)
```

```python
# ðŸ”¹ JOIN â†’ Merge tables
# Join Students and Enrollments on student_id
students_enrolled = pd.merge(students_df, enrollments_df, on='student_id')
print("\nStudents + Enrollments:\n", students_enrolled)

# Join the above with Courses on course_id
full_info = pd.merge(students_enrolled, courses_df, on='course_id')
print("\nStudents + Enrollments + Courses:\n", full_info)
```

```python
# ðŸ”¹ GROUP BY â†’ Group by Course Name
# Group by course and show mean grade
grouped = full_info.groupby('course_name')
print("\nGroupBy object:", grouped)  # Just to show it's a groupby object
```

```python
# Convert grouped to DataFrame using mean()
# Aggregate: calculate mean of all numeric columns per group
grouped_df = grouped.mean(numeric_only=True).reset_index()

# Display as a normal DataFrame
print(grouped_df)
```

```python
#ðŸ”¹ AGGREGATE â†’ Using mean(), sum(), etc.
# Average grade per course
avg_grade = full_info.groupby('course_name')['grade'].mean().reset_index()
avg_grade.columns = ['course_name', 'average_grade']
print("\nAverage Grade per Course:\n", avg_grade)

# Number of students per course
num_students = full_info.groupby('course_name')['student_id'].nunique().reset_index()
num_students.columns = ['course_name', 'num_students']
print("\nNumber of Unique Students per Course:\n", num_students)
```

```python
# Join Students and Enrollments
students_enrolled = pd.merge(students_df, enrollments_df, on='student_id')

# Join with Courses
full_info = pd.merge(students_enrolled, courses_df, on='course_id')
```

```python
# Export DataFrames to CSV
students_enrolled.to_csv('students_enrollments.csv', index=False)
full_info.to_csv('students_enrollments_courses.csv', index=False)
```
